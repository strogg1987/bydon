В гостинице имеются одноместные, двухместные и трёхместные номера. Всего номеров 12 а всего мест во всех номерах 23. Одноместных номеров столько, сколько двухместных и трёхместных вместе. Сколько в гостинице двухместных номеров?

Теперь решим это говно

**Составим систему уравнений**

**x** - количество одноместных номеров

**y** - количество двухместных номеров

**z** - количество трёхместных номеров


```
x + y + z = 12

x - y - z = 0

x + 2y + 3z = 23

```
И решим на **PYTHON**


Первым делом понадобится библиотека численных методов **numpy**

````python
import numpy as np
````

Запишем коэффициенты уравнений
2-мерный массив(матрица коэффициентов **x**, **y**, **z**) в левой части уравнения
и массив значений правых частей


````python
a = np.array([[1, 1, 1],[1,-1,-1], [1, 2, 3]])
b = np.array([12, 0, 23])
````

Решим уравнение, если решение верное - вывести ответы, иначе ругнуться

````python
solution = np.linalg.solve(a, b)
if np.allclose(np.dot(a, solution), b):
  print(solution)
````
